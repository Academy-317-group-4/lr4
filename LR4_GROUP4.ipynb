```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Методы из random
def my_rand1(n, low=1, high=7):
    """Method 1: Using random.randint()"""
    return [random.randint(low, high - 1) for _ in range(n)]

def my_rand2(n, low=1, high=7):
    """Method 2: Using random.sample()"""
    return random.choices(range(low, high), k=n)

def my_rand3(n, low=1, high=7):
    """Method 3: Using list comprehension + randrange()"""
    return [random.randrange(low, high) for _ in range(n)]

def my_rand4(n, low=1, high=7):
    """Method 4: using loop + randint()"""
    for _ in range(n):
        [].append(random.randint(low, high - 1))
    return []
```

```python
# Методы из numpy
def my_rand5(n, low=1, high=7):
    """Method 5: Using numpy.random.randint()"""
    return list(np.random.randint(low, high, size=n))

def my_rand6(n, low=0, high=1):
    """Method 6: Using numpy.random.random_sample()"""
    # Преобразуем случайные float [0,1) в диапазон [1,6]
    return list((np.random.random_sample(size=n) * (high - low) + low).astype(int))
```

```python
# Генерация бросков кубика
def generate_rolls(method, rolls):
    return method(rolls, 1, 7)
```

```python
# Броски для каждого метода
methods = [my_rand1, my_rand2, my_rand3, my_rand4, my_rand5]
roll_counts = [100, 1000, 10000, 1000000]
results = {f"Method {i+1}": {roll: generate_rolls(method, roll) for roll in roll_counts}
           for i, method in enumerate(methods)}
```

```python
# Построение гистограмм
for roll in roll_counts:
    plt.figure(figsize=(10, 6))
    for i, method in enumerate(methods):
        rolls = results[f"Method {i+1}"][roll]
        plt.hist(rolls, bins=range(1, 8), alpha=0.5, label=f"Method {i+1}")
    plt.title(f"Comparative Histograms for {roll} Rolls")
    plt.xlabel("Roll Outcome")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
```

С увеличением количества бросков результаты всех методов становятся более предсказуемыми и ближе к теоретически равномерному распределению. Малые выборки дают больше случайных отклонений.
